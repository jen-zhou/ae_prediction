```python
import matplotlib.pyplot as plt

# Count occurrences of each cancer type
cancer_counts = profile_idmap['CANCER_TYPE'].value_counts()

# Separate top 10 and group the rest as "Other"
top_10 = cancer_counts[:10]
other = cancer_counts[10:].sum()
cancer_pie = top_10.append(pd.Series({'Other': other}))

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(cancer_pie, labels=cancer_pie.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Cancer Types and Other')
plt.axis('equal')
plt.show()
```
